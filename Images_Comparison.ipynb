```python
import torch
from PIL import Image
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
import cv2

def slicing(img):
    originalImage = cv2.imread(img)
    slicedImage = originalImage[255:715, 37:688]
    
    #cv2.rectangle(originalImage,(102,80),(510,450),(0,255,0),2)
    
    #cv2.imshow("Original Image", originalImage)
    #cv2.imshow("Sliced Image", slicedImage)
    #cv2.imwrite("Sliced_Image.jpg", slicedImage)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    

def display_image(img):
    image = Image.open(img)
    #print(image.size)
    image= transforms.ToTensor()(image)  
    
    compose = transforms.Compose([transforms.ToPILImage(),transforms.Resize((128,128),interpolation=Image.NEAREST)])
    trans_img = compose(image)
    #print(trans_img.size)
    image= transforms.ToTensor()(trans_img)  
    #print(image)
    #plt.imshow(trans_img)
    return image
    


image1 = "Images/mug.jpg"
slic = "Images/mob_Img.jpeg"

slicing(slic)

image2 = "Images/Sliced_Image.jpg"
orig = display_image(image1)
mob = display_image(image2)

p_orig = transforms.ToPILImage()(orig)
p_mob = transforms.ToPILImage()(mob)


g_orig = transforms.Grayscale(num_output_channels=3)(p_orig)
plt.imshow(g_orig)
plt.show()

g_mob = transforms.Grayscale(num_output_channels=3)(p_mob)
plt.imshow(g_mob)
plt.show()

ten_orig =transforms.ToTensor()(g_orig)
ten_mob =transforms.ToTensor()(g_mob)



std_orig = torch.std(ten_orig)
std_mob = torch.std(ten_mob)


mean_val_orig=torch.mean(ten_orig)
mean_val_mob=torch.mean(ten_mob)

print("Original image Mean",mean_val_orig)
print("Original image STD",std_orig)

print("Mobile image Mean",mean_val_mob)
print("Mobile image STD",std_mob)

diff = torch.abs(ten_orig - ten_mob)
final_mean_val=torch.mean(diff)
print(final_mean_val)
```

```python
mean_diff = (final_mean_val.item()) * 100

mean = (mean_val_mob.item()) * 100

if mean >=50 and mean_diff >=15:
    print("Rejected")
else:
    print("Accepted")
```

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
import numpy as np
from PIL import ImageEnhance
from PIL import ImageFilter
import torchvision.transforms.functional as TF
import random
import imquality.brisque as brisque

   
scanned = 'scaned_bars.jpg'
original = 'bars.png'
org_img = cv2.imread(original)
sc_img = cv2.imread(scanned)

org_gray = cv2.cvtColor(org_img, cv2.COLOR_BGR2GRAY)
sc_gray = cv2.cvtColor(sc_img, cv2.COLOR_BGR2GRAY)



sc_sliced = sc_gray[320:1640, 275:1793]
p_img = transforms.ToPILImage()(sc_sliced)
r_image = transforms.Resize((480,571))(p_img)

rotated_img = TF.rotate(r_image,-1, fill=(0,))
#rotated1= cv2.cvtColor(np.float32(rotated_img), cv2.COLOR_BGR2GRAY)
rotated2 = cv2.cvtColor(np.float32(rotated_img),cv2.COLOR_GRAY2RGB)

score_scanned = brisque.score(rotated_img) 
score_org = brisque.score(org_gray)
print("Score for original Image: ",score_org)
print("Score for scanned Image: ",score_scanned)
print("--------------------------------------------")

def cal_contrast_1(img):   
    Y = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)[:,:,0]
    min = np.min(img)
    max = np.max(img)
    contrast = (max-min)/(max+min)
    return contrast

def cal_contrast(img):
    height, width = img.shape
    total_pixels = height * width
    white = cv2.countNonZero(img)
    black = total_pixels - white
    contrast = abs(white - black)
    return contrast

def edge_detect(img):
    img_gaussian = cv2.GaussianBlur(img,(3,3),0)
    kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
    kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
    img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
    return img_prewitty

def filters(img):
    org_image= transforms.ToTensor()(img) 
    composed = transforms.Compose([transforms.ToPILImage(),transforms.Resize((480,571),interpolation=Image.NEAREST)])
    trans_org_img = composed(org_image)
    

    enh_bri = ImageEnhance.Brightness(trans_org_img)
    brightness = 1.0
    image_brightened = enh_bri.enhance(brightness)

    enh_col = ImageEnhance.Color(image_brightened)
    color = 1.5
    image_colored = enh_col.enhance(color)

    enh_con = ImageEnhance.Contrast(image_colored)
    contrast = 1.0
    image_contrasted = enh_con.enhance(contrast)

    enh_sha = ImageEnhance.Sharpness(image_contrasted)
    sharpness = 1.5
    image_sharped = enh_sha.enhance(sharpness)
    return image_sharped
    
con_org = cal_contrast_1(org_img)
con_scanned = cal_contrast_1(rotated2)
print("Method 1:")
print("--------------------------------------------")
print("Contrast score for original Image: ",con_org)
print("Contrast score for scanned Image: ",con_scanned)
print("--------------------------------------------")

con_org2 = cal_contrast(org_gray)
s = transforms.ToTensor()(rotated2)
b = s[-1,:,:]
c=b.unsqueeze(-1)
print(c.shape)
con_scanned2 = cal_contrast(sc_gray)
print("Method 2:")
print("--------------------------------------------")
print("Contrast score for original Image: ",con_org2)
print("Contrast score for scanned Image: ",con_scanned2)
print("--------------------------------------------")

edge_org = edge_detect(org_gray)
edge_scanned = edge_detect(rotated2)
print("Original Image Edges/Region discontinuities:")
plt.imshow(edge_org)
plt.show()
print("Scanned Images Edges/Region discontinuities:")
plt.imshow(edge_scanned)
plt.show()

app_contrast = filters(rotated2)
plt.imshow(app_contrast)
plt.show()
```
