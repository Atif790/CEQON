```python
import torch
from PIL import Image
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
import cv2

def slicing(img):
    originalImage = cv2.imread(img)
    slicedImage = originalImage[255:715, 37:688]
    
    #cv2.rectangle(originalImage,(102,80),(510,450),(0,255,0),2)
    
    #cv2.imshow("Original Image", originalImage)
    #cv2.imshow("Sliced Image", slicedImage)
    #cv2.imwrite("Sliced_Image.jpg", slicedImage)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    

def display_image(img):
    image = Image.open(img)
    #print(image.size)
    image= transforms.ToTensor()(image)  
    
    compose = transforms.Compose([transforms.ToPILImage(),transforms.Resize((128,128),interpolation=Image.NEAREST)])
    trans_img = compose(image)
    #print(trans_img.size)
    image= transforms.ToTensor()(trans_img)  
    #print(image)
    #plt.imshow(trans_img)
    return image
    


image1 = "mug.jpg"
slic = "mob_Img.jpeg"

slicing(slic)

image2 = "Sliced_Image.jpg"
orig = display_image(image1)
mob = display_image(image2)

p_orig = transforms.ToPILImage()(orig)
p_mob = transforms.ToPILImage()(mob)


g_orig = transforms.Grayscale(num_output_channels=3)(p_orig)
plt.imshow(g_orig)
plt.show()

g_mob = transforms.Grayscale(num_output_channels=3)(p_mob)
plt.imshow(g_mob)
plt.show()

ten_orig =transforms.ToTensor()(g_orig)
ten_mob =transforms.ToTensor()(g_mob)



std_orig = torch.std(ten_orig)
std_mob = torch.std(ten_mob)


mean_val_orig=torch.mean(ten_orig)
mean_val_mob=torch.mean(ten_mob)

print("Original image Mean",mean_val_orig)
print("Original image STD",std_orig)

print("Mobile image Mean",mean_val_mob)
print("Mobile image STD",std_mob)

diff = torch.abs(ten_orig - ten_mob)
final_mean_val=torch.mean(diff)
print(final_mean_val)
```

```python
mean_diff = (final_mean_val.item()) * 100

mean = (mean_val_mob.item()) * 100

if mean >=50 and mean_diff >=15:
    print("Rejected")
else:
    print("Accepted")
```

```python
import cv2 
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

org_img = cv2.imread('mug.jpg',0)
mob_img = cv2.imread('Sliced_Image.jpg',0)

#ret,th_mob =cv2.threshold(mob_img,100,255,cv2.THRESH_BINARY)
th_org = cv2.adaptiveThreshold(org_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
th_mob = cv2.adaptiveThreshold(mob_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
composed = transforms.Compose([transforms.ToPILImage(),transforms.Resize((256,256),interpolation=Image.NEAREST)])

trans_org_img = composed(th_org)
trans_mob_img = composed(th_mob)

plt.imshow(trans_org_img,'gray')
plt.show()
plt.imshow(trans_mob_img,'gray')
plt.show()
t_org_img = transforms.ToTensor()(trans_org_img)
t_mob_img = transforms.ToTensor()(trans_mob_img)

std_org = torch.std(t_org_img)
std_mob = torch.std(t_mob_img)


mean_val_orig=torch.mean(t_org_img)
mean_val_mob=torch.mean(t_mob_img)

print("Original image Mean",mean_val_orig)
print("Original image STD",std_org)

print("Mobile image Mean",mean_val_mob)
print("Mobile image STD",std_mob)

diff = torch.abs(t_org_img - t_mob_img)
final_mean_val=torch.mean(diff)
print(final_mean_val)
```

```python
from PIL import Image
from PIL import ImageEnhance
import matplotlib.pyplot as plt

image = Image.open('Sliced_Image.jpg')
plt.imshow(image)
plt.show()

enh_bri = ImageEnhance.Brightness(image)
brightness = 1.0
image_brightened = enh_bri.enhance(brightness)

enh_col = ImageEnhance.Color(image_brightened)
color = 1.5
image_colored = enh_col.enhance(color)

enh_con = ImageEnhance.Contrast(image_colored)
contrast = 1.0
image_contrasted = enh_con.enhance(contrast)

enh_sha = ImageEnhance.Sharpness(image_contrasted)
sharpness = 2.5
image_sharped = enh_sha.enhance(sharpness)

plt.imshow(image_sharped)
plt.show()
```

```python
import imquality.brisque as brisque
from PIL import Image
from PIL import ImageEnhance
from PIL import ImageFilter
import matplotlib.pyplot as plt
import cv2

org_img = Image.open('mug.jpg')
image = Image.open('Sliced_Image.jpg')

org_image= transforms.ToTensor()(org_img) 
mob_image= transforms.ToTensor()(image) 
composed = transforms.Compose([transforms.ToPILImage(),transforms.Resize((412,512),interpolation=Image.NEAREST)])
trans_org_img = composed(org_image)
trans_mob_img = composed(mob_image)

plt.imshow(trans_org_img)
plt.show()

plt.imshow(trans_mob_img)
plt.show()

enh_bri = ImageEnhance.Brightness(trans_mob_img)
brightness = 1.0
image_brightened = enh_bri.enhance(brightness)

enh_col = ImageEnhance.Color(image_brightened)
color = 1.5
image_colored = enh_col.enhance(color)

enh_con = ImageEnhance.Contrast(image_colored)
contrast = 1.0
image_contrasted = enh_con.enhance(contrast)

enh_sha = ImageEnhance.Sharpness(image_contrasted)
sharpness = 1.5
image_sharped = enh_sha.enhance(sharpness)

#smoothenedImage = image_sharped.filter(ImageFilter.SMOOTH)

#moreSmoothenedImage = image_sharped.filter(ImageFilter.SMOOTH_MORE)

plt.imshow(image_sharped)
plt.show()

sr_org = brisque.score(trans_org_img)
sr_mob = brisque.score(trans_mob_img)
sr_mob_eff = brisque.score(image_sharped)
print("Score for original image: ",sr_org)
print("Score for Mobile Noisy image: ",sr_mob)
print("Score for Mobile Noise Removed image: ",sr_mob_eff)
```

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
import numpy as np

#img = cv2.imread('bars.png',0)
#img = cv2.imread('eins.jpg',0)
img = cv2.imread('sample_noise.jpeg',0)

 #Using CANNY Filter
canny = cv2.Canny(img, 50,100)
cv2.imshow('Canny Edge Detection', canny)


height, width = img.shape
total_pixels = height * width

white = cv2.countNonZero(img)
black = total_pixels - white

contrast = abs(white - black)
print("Contrast = ",contrast)
#print(transforms.ToTensor()(canny))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
import numpy as np

img = cv2.imread('bars.png')
#img = cv2.imread('sample_noise.jpeg')
img = cv2.imread('eins.jpg')

img_gaussian = cv2.GaussianBlur(img,(3,3),0)

# Using sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
plt.imshow(img_sobel)
```
